```python
### free parameters
paper = 1
if paper:
    ############################################
    ### assumption used in paper 1607.01009
    ############################################
    kg = 1
    year = 1
    light_yield = 1
    LCE = 1
    det_eff = 1
    DCR = 0.0
else:

    ############################################
    ### more realistic assumptions
    ############################################
    kg = 0.001
    year = 1./12
    light_yield = 0.2 #ph/eV
    LCE = 1
    det_eff = 0.01 
    DCR = 0.01 #cps
```

```python

leg = rt.TLegend(0.18,0.15,0.46,0.32)
leg.SetTextSize(0.03)
leg.SetBorderSize(0)
leg.SetEntrySeparation(0.01)

c = rt.TCanvas('c','c', 800, 800)
c.SetRightMargin(0.04)

rt.gStyle.SetOptFit(1011)
h = {}
h_exp1sig = {}
h_exp2sig = {}
h_obs = {}
h_others = {}
sample_temp = []

fdm = 0

i = 0 
for m in limits.keys():

    if fdm == 2 and not 'Fdmq-2' in m:continue
    if fdm == 0 and not 'Fdm1' in m:continue
    if "XENON" in m:continue
    print(m)
    h_others[m] = create_TGraph(limits[m][:,0], limits[m][:,1],  axis_title=['mass [MeV]', ''])
    leg.AddEntry(h_others[m],m.split('_')[-1] + '_paper', "L")
    h_others[m].SetLineColor(std_color_list[i])
    h_others[m].SetMarkerColor(std_color_list[i])

    h_others[m].SetLineWidth(3)

    h_others[m].SetLineStyle(1)
    h_others[m].SetLineWidth(3)
    h_others[m].GetXaxis().SetLimits(1,1000.0)
    h_others[m].GetYaxis().SetRangeUser(1e-43,1e-33)

    h_others[m].GetXaxis().SetTitleOffset(1)
    h_others[m].GetYaxis().SetTitleSize(0.05)
    h_others[m].GetYaxis().SetTitleOffset(1.5)
    
    i+=1


if fdm == 0:
    h['1ph'] = create_TGraph(mass/1e6, limit1ph_fdm1,  axis_title=['mass [MeV]', ''])
    h['2ph'] = create_TGraph(mass/1e6, limit2ph_fdm1,  axis_title=['mass [MeV]', ''])
else:
    h['1ph'] = create_TGraph(mass/1e6, limit1ph_fdmq2,  axis_title=['mass [MeV]', ''])
    h['2ph'] = create_TGraph(mass/1e6, limit2ph_fdmq2,  axis_title=['mass [MeV]', ''])

h['1ph'].SetLineColor(std_color_list[3])
h['2ph'].SetLineColor(std_color_list[4])

for i, m in enumerate(h.keys()):
    leg.AddEntry(h[m],m, "L")
    h[m].SetLineColor(std_color_list[i])
    h[m].SetLineWidth(3)
    h[m].SetLineStyle(2)
    h[m].SetLineWidth(3)
    h[m].GetXaxis().SetLimits(0.1,1000.0)
    if fdm == 0: h[m].GetYaxis().SetRangeUser(1e-43,1e-36)
    else: h[m].GetYaxis().SetRangeUser(1e-43,1e-33)

    h[m].GetXaxis().SetTitleOffset(1)
    h[m].GetYaxis().SetTitleSize(0.05)
    h[m].GetYaxis().SetTitleOffset(1.5)
    h[m].GetYaxis().SetTitle('#bar{#sigma_{e}} [cm^{2}]')
    h[m].SetTitle('')

for i,m in enumerate(h.keys()):
    h[m].Draw('LA' if i == 0 else 'Lsame')
for m in h_others.keys():    
    h_others[m].Draw('Lsame')
    
latex = rt.TLatex()
latex.SetTextSize(0.06)
latex.SetTextFont(42)

if fdm == 0: latex.DrawLatex(50,1e-42,  "F_{DM} = 1")
else: latex.DrawLatex(50,1e-42,  "F_{DM} ~ 1/q^{2}")





leg.Draw()
c.SetLogy()
c.SetLogx()
c.SetTicky(1)
c.SetTickx(1)
c.SetLeftMargin(0.15)
c.Draw()
```
