```python
# plot limits digitized from paper 1607.01009
import ROOT as rt
import csv
import re
import sys
import collections
import os
from collections import OrderedDict
import uproot
import pandas as pd

import scipy
import awkward
import numpy as np
import time
import numba
from numba import jit
from matplotlib import pyplot as plt
sys.path.append('/storage/af/user/christiw/login-1/christiw/DM/SNSPD_GaAs_limit/lib/')
from histo_utilities import create_TH1D, create_TH2D, std_color_list, create_TGraph

import dm_rates

from dm_rates import read_f2, d_rate, rate

f2 = read_f2('QEDark_GaAs_k8_100Ryd.npy')

```

```python
path = '/storage/af/user/christiw/login-1/christiw/DM/SNSPD_GaAs_limit/data/'

ele_recoil = {}
limits = {}

# ele_recoil['Fdmq-2_0p01GeV'] = np.genfromtxt(path + '1607.01009_Fig4_Fdmq-2_GaAs_0p01GeV.csv', delimiter = ',')
# ele_recoil['Fdmq-2_1GeV'] = np.genfromtxt(path + '1607.01009_Fig4_Fdmq-2_GaAs_1GeV.csv', delimiter = ',')
# ele_recoil['Fdm1_0p01GeV'] = np.genfromtxt(path + '1607.01009_Fig4_Fdm1_GaAs_0p01GeV.csv', delimiter = ',')
# ele_recoil['Fdm1_1GeV'] = np.genfromtxt(path + '1607.01009_Fig4_Fdm1_GaAs_1GeV.csv', delimiter = ',')


limits['Fdmq-2_1ph'] = np.genfromtxt(path + '1607.01009_Fig2_limit_Fdmq-2_GaAs_1ph.csv', delimiter = ',')
limits['Fdm1_1ph'] = np.genfromtxt(path + '1607.01009_Fig2_limit_Fdm1_GaAs_1ph.csv', delimiter = ',')
limits['Fdmq-2_2ph'] = np.genfromtxt(path + '1607.01009_Fig2_limit_Fdmq-2_GaAs_2ph.csv', delimiter = ',')
limits['Fdm1_2ph'] = np.genfromtxt(path + '1607.01009_Fig2_limit_Fdm1_GaAs_2ph.csv', delimiter = ',')

limits['Fdmq-2_XENON'] = np.genfromtxt(path + '1607.01009_Fig2_limit_Fdmq-2_XENON10_1kg-year.csv', delimiter = ',')
limits['Fdm1_XENON'] = np.genfromtxt(path + '1607.01009_Fig2_limit_Fdm1_XENON10_1kg-year.csv', delimiter = ',')
```

```python
### free parameters
paper = 0
if paper:
    ############################################
    ### assumption used in paper 1607.01009
    ############################################
    kg = 1
    year = 1
    light_yield = 1
    LCE = 1
    det_eff = 1
    DCR = 0.0
    sig_1ph = 3.6
    sig_2ph = 3.6
else:

    ############################################
    ### more realistic assumptions
    ############################################
    kg = 0.001
    year = 1./12
    light_yield = 0.2 #ph/eV
    LCE = 1
    det_eff = 0.01 
    DCR = 0.01 #cps
    window = 1e-6 # in units of second, coincidence window for 2 photons
    
    sig_1ph = 318 # can set limit with this much signal events
    sig_2ph = 1.925 # can set limit with this much signal events
```

```python
### get signal yield

signal = {}
ph_num = {}
signal_yield_1ph = {}
signal_yield_2ph = {}
bkg_1ph = {}
bkg_2ph = {}


mass = 10**np.arange(-1,3,0.1)*1e6 #conversion to eV
print(mass/1e6)
limit1ph_fdm1 = []
limit1ph_fdmq2 = []

limit2ph_fdm1 = []
limit2ph_fdmq2= []
fdm_name = ['1', '1/q^2']


for n, fdm in enumerate([0,2]):
    for l, mX in enumerate(mass):
    
        k = '{}_{}'.format(str(mX/1e6).replace('.','p'), fdm)
        E, dR = d_rate(mX, f2, fdm)
        #### paper
        if paper:
            signal_yield_1ph[k] = np.sum(dR[E > 1.52]) * dm_rates.dE
            signal_yield_2ph[k] = np.sum(dR[E> (4.2+1.52)]) * dm_rates.dE
            bkg_1ph[k] = 0.0
            bkg_2ph[k] = 0.0
            if n == 0:
                if signal_yield_1ph[k] > 0:limit1ph_fdm1.append(1/signal_yield_1ph[k]*3.6*dm_rates.crosssection)
                else: limit1ph_fdm1.append(1e6)
                if signal_yield_2ph[k] > 0:limit2ph_fdm1.append(1/signal_yield_2ph[k]*3.6*dm_rates.crosssection)
                else: limit2ph_fdm1.append(1e6)
            else:
                if signal_yield_1ph[k] > 0:limit1ph_fdmq2.append(1/signal_yield_1ph[k]*3.6*dm_rates.crosssection)
                else: limit1ph_fdmq2.append(1e6)
                if signal_yield_2ph[k] > 0:limit2ph_fdmq2.append(1/signal_yield_2ph[k]*3.6*dm_rates.crosssection)
                else: limit2ph_fdmq2.append(1e6)
        else:
            ph_num = E * light_yield
            signal_yield_1ph[k] = np.sum(dR[ph_num >= 1]) * dm_rates.dE * kg * year * LCE * det_eff
            signal_yield_2ph[k] = np.sum(dR[ph_num >= 2]) * dm_rates.dE  * kg * year * LCE* det_eff
            
           
            bkg_1ph[k] = DCR * year * 3600 * 24 * 365.24
            bkg_2ph[k] = 2 * DCR * DCR * window * year * 3600 * 24 * 365.24

            if n == 0:
                if signal_yield_1ph[k] > 0:limit1ph_fdm1.append(1/signal_yield_1ph[k]*sig_1ph*dm_rates.crosssection)
                else: limit1ph_fdm1.append(1e6)
                if signal_yield_2ph[k] > 0:limit2ph_fdm1.append(1/signal_yield_2ph[k]*sig_2ph*dm_rates.crosssection)
                else: limit2ph_fdm1.append(1e6)
            else:
                if signal_yield_1ph[k] > 0:limit1ph_fdmq2.append(1/signal_yield_1ph[k]*sig_1ph*dm_rates.crosssection)
                else: limit1ph_fdmq2.append(1e6)
                if signal_yield_2ph[k] > 0:limit2ph_fdmq2.append(1/signal_yield_2ph[k]*sig_2ph*dm_rates.crosssection)
                else: limit2ph_fdmq2.append(1e6)
    
        print(k, '\t', 1/signal_yield_1ph[k]*3.6*dm_rates.crosssection, '\t', 1/signal_yield_2ph[k]*3.6*dm_rates.crosssection, '\t', bkg_1ph[k], bkg_2ph[k])

limit1ph_fdm1 = np.array(limit1ph_fdm1)
limit1ph_fdmq2 = np.array(limit1ph_fdmq2)
limit2ph_fdm1 = np.array(limit2ph_fdm1)
limit2ph_fdmq2= np.array(limit2ph_fdmq2)
```

```python

leg = rt.TLegend(0.18,0.12,0.46,0.29)
leg.SetTextSize(0.03)
leg.SetBorderSize(0)
leg.SetEntrySeparation(0.01)

c = rt.TCanvas('c','c', 800, 800)
c.SetRightMargin(0.04)

rt.gStyle.SetOptFit(1011)
h = {}
h_exp1sig = {}
h_exp2sig = {}
h_obs = {}
h_others = {}
sample_temp = []

fdm = 2

i = 0 
for m in limits.keys():

    if fdm == 2 and not 'Fdmq-2' in m:continue
    if fdm == 0 and not 'Fdm1' in m:continue
    if paper and "XENON" in m:continue
    print(m)
    h_others[m] = create_TGraph(limits[m][:,0], limits[m][:,1],  axis_title=['mass [MeV]', ''])
    if 'XENON' in m:leg.AddEntry(h_others[m],m.split('_')[-1], "L")
    else: leg.AddEntry(h_others[m],m.split('_')[-1] + '_paper (1kg-year)', "L")
    h_others[m].SetLineColor(std_color_list[i])
    h_others[m].SetMarkerColor(std_color_list[i])

    h_others[m].SetLineWidth(3)

    h_others[m].SetLineStyle(1)
    h_others[m].SetLineWidth(3)
    h_others[m].GetXaxis().SetLimits(1,1000.0)
    h_others[m].GetYaxis().SetRangeUser(1e-43,1e-33)

    h_others[m].GetXaxis().SetTitleOffset(1)
    h_others[m].GetYaxis().SetTitleSize(0.05)
    h_others[m].GetYaxis().SetTitleOffset(1.5)
    
    i+=1


if fdm == 0:
    cond = np.logical_not(limit1ph_fdm1 == 1e6)
    h['1ph'] = create_TGraph(mass[cond]/1e6, limit1ph_fdm1[cond],  axis_title=['mass [MeV]', ''])
    cond = np.logical_not(limit2ph_fdm1 == 1e6)
    h['2ph'] = create_TGraph(mass[cond]/1e6, limit2ph_fdm1[cond],  axis_title=['mass [MeV]', ''])
else:
    cond = np.logical_not(limit1ph_fdmq2 == 1e6)
    h['1ph'] = create_TGraph(mass[cond]/1e6, limit1ph_fdmq2[cond],  axis_title=['mass [MeV]', ''])
    cond = np.logical_not(limit2ph_fdmq2 == 1e6)
    h['2ph'] = create_TGraph(mass[cond]/1e6, limit2ph_fdmq2[cond],  axis_title=['mass [MeV]', ''])

h['1ph'].SetLineColor(std_color_list[3])
h['2ph'].SetLineColor(std_color_list[4])

for i, m in enumerate(h.keys()):
#     leg.AddEntry(h[m],m + ' (1g-month)', "L")

    leg.AddEntry(h[m],m, "L")

    h[m].SetLineColor(std_color_list[i])
    h[m].SetLineWidth(3)
    h[m].SetLineStyle(2)
    h[m].SetLineWidth(3)
    h[m].GetXaxis().SetLimits(0.1,1000.0)
    if paper:
        if fdm == 0: h[m].GetYaxis().SetRangeUser(1e-43,1e-36)
        else: h[m].GetYaxis().SetRangeUser(1e-43,1e-33)
    else:
        if fdm == 0: h[m].GetYaxis().SetRangeUser(1e-45,1e-30)
        else: h[m].GetYaxis().SetRangeUser(1e-43,1e-30)
    h[m].GetXaxis().SetTitleOffset(1)
    h[m].GetYaxis().SetTitleSize(0.05)
    h[m].GetYaxis().SetTitleOffset(1.5)
    h[m].GetYaxis().SetTitle('#bar{#sigma_{e}} [cm^{2}]')
    h[m].SetTitle('')

for i,m in enumerate(h.keys()):
    h[m].Draw('LA' if i == 0 else 'Lsame')
for m in h_others.keys():    
    h_others[m].Draw('Lsame')
    
latex = rt.TLatex()
latex.SetTextSize(0.06)
latex.SetTextFont(42)

if fdm == 0: latex.DrawLatex(50,1e-42,  "F_{DM} = 1")
else: latex.DrawLatex(50,1e-42,  "F_{DM} ~ 1/q^{2}")





leg.Draw()
c.SetLogy()
c.SetLogx()
c.SetTicky(1)
c.SetTickx(1)
c.SetLeftMargin(0.15)
c.Draw()
```
